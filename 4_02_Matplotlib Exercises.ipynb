** Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
fig = plt.figure()

axe1 = fig.add_axes([0,0,1,1])
axe2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])

axe1.plot(x,x**2)
axe1.set_xlabel('X')
axe1.set_ylabel('Y')

axe2.plot(x,y)
axe2.set_xlabel('X')
axe2.set_ylabel('Y')

axe2.set_xlim([20,22])
axe2.set_ylim([30,50])
```

```python
fig = plt.figure()

axe1 = fig.add_axes([0,0,1,1])
axe2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])

#Large
axe1.plot(x,z)
axe1.set_xlabel('X')
axe1.set_ylabel('Z')

#Insert
axe2.plot(x,y)
axe2.set_title('zoom')
axe2.set_xlabel('X')
axe2.set_ylabel('Y')
axe2.set_xlim([20,22])
axe2.set_ylim([30,50])
```

## Exercise 4

** Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
```

** Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y,'b--',lw=2)

axes[1].plot(x,x**2,'r.-',lw=2)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y,'b--',lw=2)

axes[1].plot(x,z,'r.-',lw=2)
```

** See if you can resize the plot by adding the figsize() argument in plt.subplots() are copying and pasting your previous code.**

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(8,2))

axes[0].plot(x,y,'b--',lw=2)

axes[1].plot(x,x**2,'r.-',lw=2)

plt.tight_layout()
```
